```python
Rm_1 = np.array(d1["mean"])
episodes_1 = np.array(range(len(Rm_1)))
mean_rewards_1 = np.array(d1["mean"])
std_rewards_1 = np.array(d1["std"])

source1 = ColumnDataSource(dict(
    x = episodes_1,
    mean = mean_rewards_1,
    std = std_rewards_1,
    upper = mean_rewards_1 + std_rewards_1,
    lower = mean_rewards_1 - std_rewards_1
))

p1 = figure(
    title = "RL in SNN with rate neurons: FrozenLake-v0",
    x_axis_label = "Episodes",
    y_axis_label = "Rewards per step"
)

band1 = Band(base='x', lower='lower', upper='upper', source=source1, level='underlay',
            fill_alpha=0.2, line_width=1, line_color='black', fill_color='crimson')
p1.line(x=episodes_1, y=mean_rewards_1, line_width=2, color='crimson')

p1.add_layout(band1)

p1.legend.location = 'bottom_right'

show(p1)
```
